# Question 1

```python
import pandas as pd #step Libraries Imported
import numpy as np

# Step 2: Load the dataset and # Step 3 Assign it to a variable called users
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 4: Calculated the mean age per occupation
mean_age = users.groupby('occupation')['age'].mean()

# Step 5: Calculated the Male ratio per occupation
users['male'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)
male_ratio = users.groupby('occupation')['male'].mean().sort_values(ascending=False)

# Step 6: Minimum and maximum ages for each occupation
min_max_age = users.groupby('occupation')['age'].agg(['min', 'max'])

# Step 7: Mean age per occupation and gender
mean_age_gender = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: Percentage of women and men for each occupation
gender_count = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
gender_percentage = gender_count.div(gender_count.sum(axis=1), axis=0) * 100

print("Mean Age Per Occupation:\n", mean_age)
print("Male Ratio Per Occupation:\n", male_ratio)
print("Min and Max Ages:\n", min_max_age)
print("Mean Age by Occupation and Gender:\n", mean_age_gender)
print("Gender Percentage:\n", gender_percentage)
```

# Question 2

```python
# Step 1: Imported libraries
import pandas as pd

# Step 2: Imported the dataset from this address
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"

# Step 3: Assign it to a variable called euro12
euro12 = pd.read_csv(url)

# Step 4: Selected only the Goal column
goals = euro12['Goals']
print("Goals Column:\n", goals)

# Step 5: How many teams participated in the Euro 2012?
team_count = euro12['Team'].nunique()
print("\nNumber of Teams Participated:\n", team_count)

# Step 6: What is the number of columns in the dataset?
column_count = euro12.shape[1]
print("\nNumber of Columns:\n", column_count)

# Step 7: View only the columns Team, Yellow Cards, and Red Cards and assign them to a dataframe called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']].copy()
print("\nDiscipline DataFrame:\n", discipline)

# Step 8: Sorted the teams by Red Cards, then to Yellow Cards
sorted_discipline = discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending=[False, False])
print("\nSorted Discipline DataFrame:\n", sorted_discipline)

# Step 9: Calculated the mean Yellow Cards given per team
mean_yellow_cards = euro12['Yellow Cards'].mean()
print("\nMean Yellow Cards per Team:\n", mean_yellow_cards)

# Step 10: Filterd teams that scored more than 6 goals
teams_over_6_goals = euro12[euro12['Goals'] > 6]['Team']
print("\nTeams Scoring More Than 6 Goals:\n", teams_over_6_goals)

# Step 11: Selectd the teams that start with 'G'
teams_starting_with_g = euro12[euro12['Team'].str.startswith('G')]['Team']
print("\nTeams Starting with 'G':\n", teams_starting_with_g)

# Step 12: Selected the first 7 columns
first_7_columns = euro12.iloc[:, :7]
print("\nFirst 7 Columns:\n", first_7_columns)

# Step 13: Selected all columns except the last 3
all_but_last_3 = euro12.iloc[:, :-3]
print("\nAll Columns Except the Last 3:\n", all_but_last_3)

# Step 14: Presented only the Shooting Accuracy from England, Italy, and Russia
shooting_accuracy = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]
print("\nShooting Accuracy of Selected Teams:\n", shooting_accuracy)
```

```python
# Step 1: Importedlibraries
import pandas as pd
import numpy as np

# Step 2: Imported the dataset from the attached file wind.txt
# (Replace 'wind.txt' with the correct file path if it's stored locally)
file_path = 'wind.txt'
data = pd.read_csv(file_path, sep=r'\s+', header=None)

# Step 3: Assigned proper column names
data.columns = ['Yr', 'Mo', 'Dy', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']

# Step 4: Created a proper datetime index by combining Yr, Mo, and Dy columns
# Convert 'Yr', 'Mo', 'Dy' into a string in the format 'yyyy-mm-dd'
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%y-%m-%d', errors='coerce')

# Step 5: Checked if Date is being created correctly
print("First few rows with Date:\n", data[['Yr', 'Mo', 'Dy', 'Date']].head())

# Step 6: Fixed the year 2061 issue (convert 2061 to 1961)
data['Date'] = data['Date'].apply(lambda x: x.replace(year=x.year - 100) if x.year > 2000 else x)

# Step 7: Set the datetime column as the index
data = data.set_index('Date').drop(['Yr', 'Mo', 'Dy'], axis=1)

# Step 8: Ensured the Date column is of the correct datetime type
assert data.index.dtype == 'datetime64[ns]'

# Step 9: Checked the missing values
missing_values = data.isnull().sum()
print("\nMissing Values per Location:\n", missing_values)

# Step 10: Calculateed the non-missing values in the entire dataset
non_missing_total = data.notnull().sum().sum()
print("\nTotal Non-Missing Values:", non_missing_total)

# Step 11: Mean windspeeds of all locations and all times
# Use numeric_only=True to exclude non-numeric columns during the calculation
mean_windspeed = data.mean(numeric_only=True).mean()  # mean() on numeric columns only
print("\nMean Windspeed (Overall):", mean_windspeed)
```

```python
import pandas as pd

# Step 2: Imported the dataset from the provided URL
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 3: Checked the first 10 entries of the dataset
print("First 10 entries:\n", chipo.head(10))

# Step 4: Get the number of observations in the dataset
num_observations = chipo.shape[0]
print("\nNumber of observations:", num_observations)

# Step 5: Get the number of columns in the dataset
num_columns = chipo.shape[1]
print("\nNumber of columns:", num_columns)

# Step 6: Print the name of all the columns
print("\nColumns in the dataset:", chipo.columns.tolist())

# Step 7: Checked how the dataset is indexed
print("\nDataset index:\n", chipo.index)

# Step 8: Found the most-ordered item
most_ordered_item = chipo.groupby('item_name')['quantity'].sum().idxmax()
print("\nMost ordered item:", most_ordered_item)

# Step 9: Get the number of items ordered for the most ordered item
most_ordered_quantity = chipo.groupby('item_name')['quantity'].sum().max()
print("\nQuantity of the most ordered item:", most_ordered_quantity)

# Step 10: Most ordered item in the choice_description column
most_ordered_choice_description = chipo.groupby('choice_description')['quantity'].sum().idxmax()
print("\nMost ordered choice description:", most_ordered_choice_description)

# Step 11: Total number of items ordered in the dataset
total_items_ordered = chipo['quantity'].sum()
print("\nTotal items ordered:", total_items_ordered)

# Step 12: Convert item price to a float (removing "$" sign and converting to float)
chipo['item_price'] = chipo['item_price'].replace({r'\$': ''}, regex=True).astype(float)

# Step 13: Check the item price type
print("\nItem price type after conversion:", chipo['item_price'].dtype)

# Step 14: Create a lambda function to convert item_price type
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x))

# Step 15: Check the item price type again after using lambda
print("\nItem price type after applying lambda:", chipo['item_price'].dtype)

# Step 16: Calculate the total revenue for the period in the dataset
revenue = (chipo['item_price'] * chipo['quantity']).sum()
print("\nTotal revenue in the dataset: $", revenue)

# Step 17: Total number of orders made in the period
num_orders = chipo['order_id'].nunique()
print("\nNumber of orders in the period:", num_orders)

# Step 18: Calculate the average revenue per order
avg_revenue_per_order = revenue / num_orders
print("\nAverage revenue per order: $", avg_revenue_per_order)

# Step 19: Calculate how many different items are sold
num_items_sold = chipo['item_name'].nunique()
print("\nNumber of different items sold:", num_items_sold)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: the CSV file
file_path = 'us-marriages-divorces-1867-2014.csv'  
data = pd.read_csv(file_path)

# Step 3: Calculate 'Marriages_per_capita' and 'Divorces_per_capita'
data['Marriages_per_capita'] = (data['Marriages'] / data['Population']) * 1000
data['Divorces_per_capita'] = (data['Divorces'] / data['Population']) * 1000

# Step 4: Plot the data
plt.figure(figsize=(10, 6))

# Plot marriages per capita
plt.plot(data['Year'], data['Marriages_per_capita'], label='Marriages per Capita', color='Green', linestyle='-', marker='o')

# Plot divorces per capita
plt.plot(data['Year'], data['Divorces_per_capita'], label='Divorces per Capita', color='Orange', linestyle='--', marker='x')

# Step 5: Label the axes and add a title
plt.xlabel('Year')
plt.ylabel('Per Capita (per 1000 people)')
plt.title('Marriages and Divorces Per Capita in the U.S. (1867-2014)')

# Step 6: Show the legend
plt.legend()

# Step 7: Show the plot
plt.grid(True)
plt.tight_layout()  # Ensures that the plot fits within the figure area
plt.show()
```

# Question 7

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the data from the CSV file
file_path = 'us-marriages-divorces-1867-2014.csv'  
data = pd.read_csv(file_path)

# Step 3: Calculate 'Marriages_per_capita' and 'Divorces_per_capita'
data['Marriages_per_capita'] = (data['Marriages'] / data['Population']) * 1000
data['Divorces_per_capita'] = (data['Divorces'] / data['Population']) * 1000

# Step 4: Filter the data for the years 1900, 1950, and 2000
years_of_interest = [1900, 1950, 2000]
data_filtered = data[data['Year'].isin(years_of_interest)]

# Step 5: Create a vertical bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the data
bar_width = 0.35  # Width of the bars
index = range(len(data_filtered))  # Bar positions

# Marriages per capita (blue bars)
ax.bar(index, data_filtered['Marriages_per_capita'], bar_width, label='Marriages per Capita', color='Yellow')

# Divorces per capita (red bars), offset by the bar width
ax.bar([p + bar_width for p in index], data_filtered['Divorces_per_capita'], bar_width, label='Divorces per Capita', color='Blue')

# Step 6: Label the axes and add a title
ax.set_xlabel('Year')
ax.set_ylabel('Per Capita (per 1000 people)')
ax.set_title('Marriages and Divorces Per Capita in the U.S. (1900, 1950, 2000)')
ax.set_xticks([p + bar_width / 2 for p in index])  # Place x-ticks in between the bars
ax.set_xticklabels(data_filtered['Year'])  # Use the years as the x-tick labels

# Step 7: Add legend
ax.legend()

# Show the plot
plt.tight_layout()  # Ensure the plot fits within the figure area
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2:  the CSV file
file_path = 'actor_kill_counts.csv'  
data = pd.read_csv(file_path)


# Step 4: Sort the actors by their kill count (use 'Count' column instead of 'Kill Count')
data_sorted = data.sort_values('Count', ascending=False)

# Step 5: Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(data_sorted['Actor'], data_sorted['Count'], color='black')

# Step 6: Label the axes and add a title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood (By Kill Count)')

# Add the kill count labels on the bars
for index, value in enumerate(data_sorted['Count']):
    plt.text(value, index, str(value), va='center', ha='left')

# Show the plot
plt.tight_layout()  # Ensures that the plot fits within the figure area
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the data from the CSV file
file_path = 'roman-emperor-reigns.csv' 
data = pd.read_csv(file_path)

# Step 4: "Assassinated": Any entry in the Cause_of_Death column that specifically contains the word "Assassinated" and "Not Assassinated": Any entry that does not contain "Assassinated", including values like "Natural causes", "Illness", "Suicide", etc.
# We will treat 'Possibly assassinated' and 'Assassinated' as assassinated
assassinated_count = data['Cause_of_Death'].str.contains('Assassinated', case=False, na=False).sum()

# Step 5: Calculate the number of non-assassinated emperors
non_assassinated_count = len(data) - assassinated_count

# Data for the pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [assassinated_count, non_assassinated_count]
colors = ['salmon', 'lightgreen']

# Step 6: Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

# Ensure the pie chart is an even circle
plt.axis('equal')

# Add a title
plt.title('Fraction of Roman Emperors Assassinated')

# Show the plot
plt.show()
```
